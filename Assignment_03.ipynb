List: Lists are used to store multiple items in a single variable.

Lists are one of 4 built-in data types in Python used to store collections of data, the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.


```python
#Example: 
thislist = ["apple", "banana", "cherry"]
print(thislist)
```

```python
#List Items: List items are ordered, Changeable, and allow duplicate values;
#Ordered: When we say that list are ordered, it means that the items have a defined order, and that order will not change.
#Changeable: The list is changeable, meaning that we can change, add, and remove items in a list after it has create. 
#Allow Duplicates: In a list there are one or more items with same value;
#Example: 
first_list = ["samsung", "apple", "realme", "nokia", "oppo", "apple"]
print(first_list)
```

```python
#List Length: To determine how many items in a list, we use the len() function;
print(len(first_list));
```

```python
#The list() Constructor: For creating a list we use the constructor of list()
second_list = list(('apple','samsung', 'nokia', 'walton', 'oppo'));
print(second_list);
```

```python
#Access Items: list item are decorated by their indexed, for access item from a list we call by their index number;
print(second_list[2]);
```

```python
# we also call the negative index number
print(second_list[-2]);
```

```python
#range of list : we can call multiple items from a list by calling their range of index;
print(second_list[2:4]); #print from to index to 4(not 4, before 4)
```

```python
second_list = list(('apple','samsung', 'nokia', 'walton', 'oppo'))
print(second_list[:3]) #print form first to before 3;
```

```python
print(second_list[2:]) # print 2 to last
```

```python
#check it a item exists:
if 'oppo' in second_list:
    print('yes');
```

```python
second_list = list(('apple','samsung', 'nokia', 'walton', 'oppo'))
second_list[1] = 'motorola'; #exchange the value of index 1;
print(second_list);
```

```python
second_list[1:3] = ['samsung', 'apple', 'walton' ];
print(second_list);
```

```python
thislist = ["apple", "banana", "cherry"]
thislist[1:2] = ["blackcurrant", "watermelon"]
print(thislist)
```
